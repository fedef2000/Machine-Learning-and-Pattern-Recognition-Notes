# Function to load data

```python
import csv, numpy, matplotlib.pyplot as plt
```

```python
def loadValues(filename):
    data = []
    labels = []
    # Open the CSV file
    with open(filename) as file:
        reader = csv.reader(file)
        # Loop through rows
        for row in reader:
            data.append(row[0:4])
            if(row[4]=="Iris-setosa"): labels.append(0)
            elif (row[4]=="Iris-versicolor"): labels.append(1)
            elif (row[4]=="Iris-virginica"): labels.append(2)

    return numpy.array(data, dtype=numpy.float64).T, numpy.array(labels, dtype=numpy.float64)

data, label = loadValues("iris.csv")

#print("first element: " + str(data[:, 0]) + ", label: " + str(label[0]))
```

filter data based on label

```python
data0 = data[:, label==0]
data1 = data[:, label==1]
data2 = data[:, label==2]
```

# Histograms

Note: density parameter is used for normalize the data. It allows you to compare histograms of different classes, even if they have different sample sizes and it makes the histogram independent of the total number of data points. In the histogram each rectangle is divided by the total number of samples.

```python

plt.figure()
plt.hist(data0[0], density = True, alpha=0.4)
plt.hist(data1[0], density = True, alpha=0.4)
plt.hist(data2[0], density = True, alpha=0.4)
plt.show()
```
