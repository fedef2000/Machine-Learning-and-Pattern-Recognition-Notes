```python
import csv, numpy, matplotlib.pyplot as plt
```

```python
def loadValues(filename):
    data = []
    labels = []
    # Open the CSV file
    with open(filename) as file:
        reader = csv.reader(file)
        # Loop through rows
        for row in reader:
            data.append(row[0:6])
            labels.append(row[6])

    return numpy.array(data, dtype=numpy.float64).T, numpy.array(labels, dtype=numpy.float64)

data, label = loadValues("33880408.txt")

print("first element: " + str(data[:, 0]) + ", label: " + str(label[0]))
```

```python
fake = data[:, label == 0]
true = data[:, label == 1]
```

## Question 1
Analyze the first two features. What do you observe? Do the classes overlap? If so, where? Do the
classes show similar mean for the first two features? Are the variances similar for the two classes?
How many modes are evident from the histograms (i.e., how many “peaks” can be observed)?

```python
plt.Figure()
plt.hist(fake[0], alpha=0.5, label="Fake", density=True)
plt.hist(true[0], alpha=0.5, label="True",  density=True)
plt.legend(loc='upper right')
plt.show()
```

```python
plt.Figure()
plt.hist(fake[1], alpha=0.5, label="Fake", density=True)
plt.hist(true[1], alpha=0.5, label="True", density=True)
plt.legend(loc='upper right')
plt.show()
```

```python
print("true mean: " + str(true.mean(axis=1)[0:2]))
print("fake mean: " + str(fake.mean(axis=1)[0:2]))
print("true variance: " + str(true.var(axis=1)[0:2]))
print("fake variance: " + str(fake.var(axis=1)[0:2]))
```

Answer: Yes, they both overlap in the middle (that is also the mean). They both have a normal distrubution around the same mean. The variance of feature 1 true is really similar to the variance of feature 2 fake, while the variance of feature 1 fake is really similar to the variance of feature 2 true.
